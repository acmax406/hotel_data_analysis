```python
import pandas as pd
import seaborn as sns
import numpy as np
import os
```

```python
df= pd.read_csv("reviews_kualalampur.csv")
df.columns=["rating","time","review","comment","hotel_id","hotel_name"]
df.to_csv("temp.csv")
df
```

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull(),yticklabels=False)
```

```python
df=df.dropna(subset=['review'])
df['hotel_name'] = df['hotel_name'].str.replace('\n','')
df['hotel_name'] = df['hotel_name'].str.replace('Opens in new window','')
df
```

```python
new = df["comment"].str.split("Disliked\n .", n = 4, expand = True)
df["comment1"]= new[0]
df["comment2"]= new[1]
# df.to_csv("temp.csv",index=False)
df['comment1'] = df['comment1'].str.replace('\n','')
df['comment1'] = df['comment1'].str.replace('Liked .','')
df.head()
```
