```python
import numpy as np
import pandas as pd
import seaborn as sns 
import os
```

```python
missing_value=["N/a","na",np.nan]
df = pd.read_csv("ratings_kualalampur.csv",na_values=missing_value)
# df.iloc[:,:1]
df.columns=['staff_r','facilities_r','cleanliness_r','comfort_r','Value_for_money_r','location_r','Free_wifi','overall','hotel_name','value']
df
```

```python
#to check how many null values are there in the columns
df.isnull().sum()
```

```python
#to check the null values in the columns
# df.isnull().any()
#heatmap to find the null values in columns
sns.heatmap(df.isnull(),yticklabels=False)
```

```python
df=df.dropna()
df
```

```python
df['hotel_name'] = df['hotel_name'].str.replace('\n','')
df['hotel_name'] = df['hotel_name'].str.replace('Opens in new window','')
# df['hotel_name'] = df['hotel_name'].str.replace('-','')
# df['hotel_name'] = df['hotel_name'].str.replace('SHA Extra Plus','')
df.to_csv("kualalampur_c_r.csv", index =False)
```

```python
from geopy.exc import GeocoderTimedOut
from geopy.geocoders import Nominatim

longitude = []
latitude = []

def findGeocode(hotel_name):
	try:
		geolocator = Nominatim(user_agent="your_app_name")
		return geolocator.geocode(hotel_name)
	except GeocoderTimedOut:		
		return findGeocode(hotel_name)	

for i in (df["hotel_name"]):	
	if findGeocode(i) != None:		
		loc = findGeocode(i)
		latitude.append(loc.latitude)
		longitude.append(loc.longitude)
	else:
		latitude.append(np.nan)
		longitude.append(np.nan)
```

```python
df["Longitude"] = longitude
df["Latitude"] = latitude
  
df
```

```python
sns.heatmap(df.isnull(),yticklabels=False)

```
