```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bangkok_data = pd.read_csv("ratings_bangkok.csv")
bangkok_data["city"] = "Bangkok"
kulalampur_data= pd.read_csv("ratings_kulalampur.csv").iloc[:, 1:]
kulalampur_data["city"] = "Kulalampur"
singapore_data = pd.read_csv("ratings_singapore.csv")
singapore_data["city"] = "Singapore"

ratings_data=pd.concat([bangkok_data, kulalampur_data, singapore_data])
```

```python
ratings_data.to_csv("data/ratings_data.csv", index=False)
```

```python
ratings_data = ratings_data[['value', 'city', 'hotel_name','overall', 'staff_r', 'facilities_r', 'cleanliness_r', 'comfort_r',
       'Value_for_money_r', 'location_r', 'Free_wifi',  
       ]]

ratings_data.rename(columns={'value':"hotel_id"}).to_csv("data/rating_data_final.csv", index=False)
```

```python
sns.regplot(data=ratings_data, x="location_r", y="overall", ci=None)
```
